```python
import pandas as pd
data={
    "Name":["Pranav","Yash","Pratik","Rohan","Virat","Rohit","Mohit"],
    "Age":[21,22,23,24,25,26,27],
    "Salary":[50000,60000,70000,80000,45000,100000,55555],
    "Performance Score":[85,74,92,66,55,78,95]
}
df=pd.DataFrame(data)
df

high_salary =df[df['Salary']>50000]
print(high_salary)

filtered=df[(df["Age"]>20) & (df["Salary"]>50000)]
print(filtered)

filtered_or=df[(df["Age"]>20) | (df["Salary"]>50000)]
print(filtered_or)
```

```python
"""Linear , Polynomial , Time"""


import pandas as pd
data={
    "Time":[1,2,3,4,5],
    "Value":[10,None,30,None,50]
}
df=pd.DataFrame(data)
print("Before Interpolation")
print(df)

df["Value"]=df["Value"].interpolate(method="linear")
print("After Interpoliation")
print(df)
```

```python
#Sorting data
#df.sort_values(by="column_name",True/False,inplace=True)

import pandas as pd
data={
    "Name":["Pranav","Yash","Pratik","Rohan","Virat","Rohit"],
    "Age":[25,22,23,26,21,24],
    "Salary":[50000,60000,70000,80000,45000,100000]
}
df=pd.DataFrame(data)
print(df)

df.sort_values(by="Age",ascending=True,inplace=True)
print("\nSorted Age by Descending")
print(df)
```

```python
#Agreegation
#df["column name"].mean(),max(),min(),sum(),count(),std()

import pandas as pd
data={
    "Name":["Pranav","Yash","Pratik","Rohan","Virat","Rohit"],
    "Age":[25,22,23,26,21,24],
    "Salary":[50000,60000,70000,80000,45000,100000]
}
df=pd.DataFrame(data)
print(df)

avg_sal=df["Salary"].mean()
print(avg_sal)

avg_sal=df["Salary"].max()
print(avg_sal)

avg_sal=df["Salary"].count()
print(avg_sal)
```

```python
#Grouping
import pandas as pd
data={
    "Name":["Pranav","Yash","Pratik","Rohan","Virat"],
    "Age":[28,34,22,34,28],
    "Salary":[50000,60000,45000,52000,480000]
}
df=pd.DataFrame(data)
print(df)

grouped=df.groupby("Age")["Salary"].sum()
print(grouped)
```

```python
#df.merge[df1,df2,on="Column_name",how="type of join")

import pandas as pd

df_customers=pd.DataFrame({
    "Name":["Pranav","Yash","Pratik"],
    "Customer id":[1,2,3]
})

df_orders=pd.DataFrame({
    "Customer id":[1,2,4],
    "Order amount":[120,250,300]
})

df_merge=pd.merge(df_customers,df_orders,on="Customer id",how="outer")
print("Inner join")
print(df_merge)
```
